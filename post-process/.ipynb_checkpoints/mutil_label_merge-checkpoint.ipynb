```python
f_in = '/public/lixin/nii/'
f_o = '/public/lixin/DATA/segthor/out_summary/out-3-22/pre-nii/'

file_pink = os.listdir(f_in)

print(file_pink)
for f in file_pink:

    print(os.path.join(f_o, f))
    file_o = sitk.GetArrayFromImage(sitk.ReadImage(os.path.join(f_o, f)))
    file_i = np.round(sitk.GetArrayFromImage(sitk.ReadImage(os.path.join(f_in, f))))
    print(np.shape(file_i))
    print(os.path.join(f_in, f))
    
    file_o[np.equal(file_o, 1)] = 0
    file_o[np.equal(file_i, 1)] = 1


    
    ori_path = os.path.join(f_o, f)
    
    print(np.shape(file_o))
    
    print(ori_path)
    
    o_folder = 'result/'
    out_path = os.path.join(o_folder, f)
    
    print(out_path)
    WriteNiiFromArray(file_o, out_path, **GetSettings(ori_path))
    
    
#     np.save(os.path.join(o_folder, f), out)
```
